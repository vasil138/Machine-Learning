1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.Symbol('x')
u = 10 * x - 5
```

```python
def diff_f_lin(f):
    df = sp.diff(f, x)

    print("Похідна лінійної функції:", df)

    f_numeric = sp.lambdify(x, f, 'numpy')
    df_numeric = sp.lambdify(x, df, 'numpy')

    x_vals = np.linspace(0, 10, 20)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, f_numeric(x_vals), label="Графік функції", color='blue')
    plt.plot(x_vals, [df_numeric(0)] * len(x_vals), label="Графік похідної", color='green')
    plt.title("Графік функції та її похідної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
u = 10 * x - 5
```

```python
diff_f_lin(u)
```

Дивлячись на графік, можна зробити висновок, що оцінка прямопропорційно залежить від витраченного часу.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
def diff_f(f):
    df = sp.diff(f, x)

    print("Похідна функції:", df)

    f_numeric = sp.lambdify(x, f, 'numpy')
    df_numeric = sp.lambdify(x, df, 'numpy')

    x_vals = np.linspace(0, 10, 100)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, f_numeric(x_vals), label="Графік функції", color='blue')
    plt.plot(x_vals, df_numeric(x_vals), label="Графік похідної", color='green')
    plt.title("Графік функції та її похідної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
v = sp.sin(x)
```

```python
diff_f(v)
```

```python
l = sp.ln(x)
```

```python
diff_f(l)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
cost = x**2 - 5*x + 1
```

```python
def diff_cost(f):
    df = sp.diff(f, x)

    print("Похідна функції:", df)

    f_numeric = sp.lambdify(x, f, 'numpy')
    df_numeric = sp.lambdify(x, df, 'numpy')

    x_vals = np.linspace(0, 100, 100)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, f_numeric(x_vals), label="Графік функції", color='blue')
    plt.plot(x_vals, df_numeric(x_vals), label="Графік похідної", color='green')
    plt.title("Графік функції та її похідної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
diff_cost(cost)
```
